# Training Pipeline
## Mendesain Model

Training pipeline adalah serangkaian langkah atau proses yang diperlukan untuk melatih model dalam machine learning atau deep learning. Ini adalah alur kerja yang mencakup semua tahap, dari mempersiapkan data hingga menghasilkan model yang terlatih. Training pipeline sangat penting dalam pengembangan model machine learning yang efektif dan dapat diandalkan.

```python
# Mengimpor Library
import torch
import torch.nn as nn

X = torch.tensor([[1], [2], [3], [4]], dtype=torch.float32)
Y = torch.tensor([[2], [4], [6], [8]], dtype=torch.float32)

n_samples, n_features = X.shape
print(n_samples, n_features)

input_size = n_features
output_size = n_features
```

```python
# Mendesain Model
model = nn.Linear(input_size, output_size)

X_test = torch.tensor([[25]], dtype=torch.float32)
print(f'Prediksi Awal: {model(X_test).item()}')
```

## Merancang loss dan optimizer

```python
def forward(x):
    return x * w


# Menggunakan loss function dari torch
loss = nn.MSELoss()     # Mean Squared Error Loss

# Menggunakan optimizer dari torch
optimizer = torch.optim.SGD(model.parameters(), lr=0.001)   # Stochastic Gradient Descent optimizer
```

## Training Loop

```python
n_iters = 10000

for epoch in range(n_iters):
    # forward pass
    y_pred = model(X)

    # backward pass
    l = loss(Y, y_pred)

    # gradients
    l.backward()

    # memperbaharui weights
    optimizer.step()

    # menghapus gradient
    optimizer.zero_grad()

    # menyimpan bias dan weights
    [w, b] = model.parameters()

    # mencetak nilai weight, bias, dan loss dengan progress bar
    if epoch % 50 == 0:
        print(f'Epoch {epoch+1}/{n_iters}')
        print(f'Weight: {w.item()}')
        print(f'Bias: {b.item()}')
        print(f'Loss: {l.item()}')
        print()
```

```python
# prediksi setelah training
print(f'Prediksi Akhir: {model(X_test).item()}')
```
